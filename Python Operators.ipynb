```python
# 1. Calculate the sum, difference, product, and quotient of two numbers.

num1 = 24
num2 = 12

print(f"Sum of num1 and num2 is {num1 + num2}")
print(f"Difference of num1 and num2 is {num1 - num2}")
print(f"Product of num1 and num2 is {num1 * num2}")
print(f"Quotient of num1 and num2 is {num1 / num2}")
```

```python
# 2. Perform various assignment operations on a variable.

num = 45    # assignment
print(num)

num += 45  # addition assignment
print(num)

num -= 45   # subraction assignment
print(num)

num *= 45   # multiplication assignment
print(num)

num /= 45    # division assignment
print(num)
```

```python
# 3. Compare two numbers and print the results.


print(f"Equal to: {23 == 23}")
print(f'Not equal to: {23 != 23}')
print(f"Greater than: {20 > 11}")
print(f"Less than: {20 < 10}")
print(f"Greater than or equal to: {20 >= 20}")
print(f"Less than or equal to: {10 <= 10}")
```

```python
# 4. Check conditions using logical operators.

logical_And = (13 > 10)  and (20 < 25)
print(logical_And)

logical_Or = (13 < 10)  or  (25 > 23)
print(logical_Or)

logical_Not = not(10 < 13)
print(logical_Not)
```

```python
# 5. Check the identity of variables.

name1 = "identity"
name2 = "pwskills"

result = name1 is name2
print(result)

result2 = name1 is not name2
print(result2)
```

```python
# 6. Perform bitwise operations on any two integers.

number1 = 7 #111
number2 = 10 #1010

Bit_And = number1 & number2
print(f"And: {Bit_And}")

Bit_Or = number1 | number2
print(f"Or: {Bit_Or}")

Bit_Not = ~ number1
print(f"Not: {Bit_Not}")

Bit_XOR = number1 ^ number2
print(f"Xor: {Bit_XOR}")

leftshift = number1 << 1
print(f"Leftshift: {leftshift}")

rightshift = number2 >> 1
print(f"Rightshift: {rightshift}")
```

```python
# 7. Use unary operators to change the sign of a number.

num = 24
print(+num)
print(-num)
print(~num)
```

```python
# 8. Use the ternary operator to assign values based on conditions.

a = 20
b = 25

compare = "a is less than b" if a < b else "a is greater than b"
print(compare)
```
